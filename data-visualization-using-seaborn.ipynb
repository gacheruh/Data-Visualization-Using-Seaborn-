
## Data Visualization Techniques


### a) Lineplot

To visualize the trends of the COVID-19 variables, we employ a line plot. The line plot displays the values of the selected variables along the y-axis and the dates along the x-axis.

By plotting the "Confirmed" cases, "Recovered" cases, "Active" cases, and "Deaths" on the same graph, we can observe the changes in these variables over time and compare their trajectories.

The line plot allows us to identify any significant increases or decreases in the number of confirmed cases, recoveries, active cases, or deaths. It also enables us to analyze the relative trends and patterns among these variables.

Through the line plot, we gain insights into the progression of the COVID-19 pandemic and the impact of various measures taken to control its spread. It helps us understand the effectiveness of efforts to curb the virus and track the recovery rate.



```python
# Create a lineplot
sns.set_style("whitegrid")
plt.figure(figsize = (18,6))
plt.title("Covid-19 by day", fontsize=22, fontweight = "bold")

sns.lineplot(data = my_data['Confirmed'], label = "Confirmed")
sns.lineplot(data = my_data['Recovered'], label = "Recovered")
sns.lineplot(data = my_data['Active'], label = "Active")
sns.lineplot(data = my_data['Deaths'], label = "Deaths")
plt.xlabel('Date', fontsize=22, fontweight = "bold")
plt.ylabel('Number of Cases', fontsize=22, fontweight = "bold")
plt.title("Trends of Covid-19 cases", fontsize=22, fontweight = "bold")
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.legend()
plt.show()




# Check that a figure appears below
step_4.check()
```

### b) Stacked Area Plot:

By creating a stacked area plot, we are able to visualize the cumulative values of variables over time. By plotting the variables "Confirmed," "Deaths," and "Recovered" as stacked areas, we can observe the overall distribution and proportions of each variable as they evolve over the course of the pandemic.



```python
plt.figure(figsize=(12, 6))
plt.stackplot(my_data.index, my_data['Confirmed'], my_data['Deaths'], my_data['Recovered'],
              labels=['Confirmed', 'Deaths', 'Recovered'])
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.title('Stacked Area Plot of COVID-19 Cases')
plt.legend()
plt.show()
```

### c) Pie Chart


Next, we use a pie chart to display the proportion of variables as a percentage of the whole. We can create a pie chart to visualize the proportions of "Deaths / 100 Cases," "Recovered / 100 Cases," and "Deaths / 100 Recovered" for a specific date. This helps us understand the relative impact of these metrics in the context of the total number of cases.


```python
# Select a specific date
selected_date = '2020-01-22'
selected_row = my_data.loc[selected_date]

labels = ['Deaths / 100 Cases', 'Recovered / 100 Cases', 'Deaths / 100 Recovered']
sizes = [selected_row['Deaths / 100 Cases'], selected_row['Recovered / 100 Cases'], selected_row['Deaths / 100 Recovered']]

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title(f'Proportions on {selected_date}')
plt.axis('equal')
plt.show()
```

### d) Scatter Plot


Scatter Plot
To explore the relationship between two variables, we employ a scatter plot. Specifically, we examine the relationship between "Confirmed" cases and either "Deaths" or "Recovered" cases. The scatter plot allows us to identify any potential correlation or patterns between these variables.



```python
plt.figure(figsize=(10, 6))
plt.scatter(my_data['Confirmed'], my_data['Deaths'], label='Deaths')
plt.scatter(my_data['Confirmed'], my_data['Recovered'], label='Recovered')
plt.xlabel('Confirmed Cases')
plt.ylabel('Number of Cases')
plt.title('Scatter Plot of COVID-19 Cases')
plt.legend()
plt.show()
```

### e) Heatmap

Heatmap
Lastly, we utilize a heatmap to visualize the correlation between different variables in the dataset. By creating a correlation matrix and plotting it as a heatmap, we can identify the strength and direction of the relationships between variables. This helps us gain insights into how different factors are interrelated within the COVID-19 dataset.



```python
plt.figure(figsize=(10, 6))
sns.heatmap(my_data.corr(), annot=True)
plt.title('Correlation Heatmap of COVID-19 Variables')
plt.show()
```
